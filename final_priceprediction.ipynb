```python
import joblib
import pandas as pd

# Load your saved pipeline
pipeline = joblib.load('midcrop_price_model.joblib')
pipelinemin = joblib.load('mincrop_price_model.joblib')
pipelinemax = joblib.load('maxcrop_price_model.joblib')

def get_user_input():
    """Gets user input for model prediction."""
    try:
        state = input("Enter State: ")
        commodity = input("Enter Commodity: ")
        year = int(input("Enter Year: "))
        month = int(input("Enter Month: "))
        rainfall = float(input("Enter Rainfall: "))
        max_temp = float(input("Enter Max Temp: "))
        min_temp = float(input("Enter Min Temp: "))
        soil = input("Enter Soil Type: ")
        humidity = float(input("Enter Humidity: "))

        return {
            'State': state,
            'Commodity': commodity,
            'Year': year,
            'Month': month,
            'Rainfall': rainfall,
            'Max_Temp': max_temp,
            'Min_Temp': min_temp,
            'Predominant Soil Types': soil,
            'Humidity': humidity
        }
    except ValueError:
        print("Invalid input. Please enter valid values.")
        return None

def predict_price(input_data):
    """Predicts crop price using the model."""
    try:
        input_df = pd.DataFrame([input_data])

        # Convert categorical columns to 'category' type
        for col in ['State', 'Commodity', 'Predominant Soil Types']:
            input_df[col] = input_df[col].astype('category')

        prediction = pipeline.predict(input_df)[0]
        print(f"Predicted Mid Price: {prediction:.2f}")
        predictionmin = pipelinemin.predict(input_df)[0]
        print(f"Predicted Min Price: {predictionmin:.2f}")
        predictionmax = pipelinemax.predict(input_df)[0]
        print(f"Predicted Max Price: {predictionmax:.2f}")
    except Exception as e:
        print(f"Error during prediction: {e}")

# Main interaction loop
while True:
    print("\nEnter input data for prediction (or 'q' to quit):")
    user_input = get_user_input()

    if user_input:
        predict_price(user_input)

    if input("Continue? (y/n): ").lower() != 'y':
        break

print("Exiting...")
```
